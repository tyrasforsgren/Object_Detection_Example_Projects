```python
from media_classifier_adapter import ImageClassification
from media_classifier_adapter import ImageToVideo_ClassificationAdapter
from media_classifier_adapter import VideoClassification
import cv2
import matplotlib.pyplot as plt
```

```python
def display(img,cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
image_path = "meeting.jpg"
image = cv2.imread(image_path)

# Create an instance of ImageClassification
image_classifier = ImageClassification(image_path)
```

```python
display(image)
```

```python
image_classifier.gather_image_info() # Find objects n store in self.image_info
image_classifier.image_info
```

```python
boxed_image = image_classifier.draw_boxes_on_image() # Createbox img
image_classifier.display_image(boxed_image)
```

```python
image_classifier.save_boxes_to_folders()
```

```python
# Vid path
video_path = "short_yoga.mp4"

# Create an instance of ImageClassification to be used by VideoClassification
image_classifier = ImageClassification(video_path)
adapter = ImageToVideo_ClassificationAdapter(video_path,image_classifier)
```

```python
frames = adapter.get_all_frames_from_vid()
len(frames),type(frames[0])
```

```python
adapter.gather_video_info(frames) # Get info on all obj for all frames
```

```python
adapter.all_frame_obj_info
```

```python
video_classifier = VideoClassification(video_path, image_classifier)
```

```python
box_frames = video_classifier.box_video_frames(frames)
```

```python
image_classifier.display_image(box_frames[0])
```

```python
image_classifier.save_boxes_to_folders()
```
